```python
x = df.drop(columns=['price_range'])
y = df['price_range']
```

```python
import random

def split_data(x, y, test_size):
    data = list(zip(x, y))
    random.shuffle(data)
    split_index = int(len(data) * test_size)
    x_train = [item[0] for item in data[split_index:]]
    y_train = [item[1] for item in data[split_index:]]
    x_test = [item[0] for item in data[:split_index]]
    y_test = [item[1] for item in data[:split_index]]
    return x_train, x_test, y_train, y_test
x_train, x_test, y_train, y_test = split_data(x, y, test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
```

```python
knn.score(x_train, y_train)
```

```python
predictions = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , predictions)
```
